```python
import pandas as pd
import geopandas as gpd
from geopandas import GeoDataFrame

# for mapping
import folium
from shapely.geometry import Point, Polygon
from keplergl import KeplerGl
```

```python
durham_geo = gpd.read_file("geodata/2016_Census_-_Durham_Region_Dissemination_Areas.shp")
```

```python
durham_geo = durham_geo.to_crs(4326)
durham_geo['lon'] = durham_geo.centroid.x  
durham_geo['lat'] = durham_geo.centroid.y
```
